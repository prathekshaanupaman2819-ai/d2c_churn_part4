вЂ‹1. Exploratory Analysis Framework

вЂ‹Customer Demographics: Analyze age, location, and account tenure to identify if specific profiles correlate with higher attrition.

вЂ‹Order & Monetary Behaviour:
 Segment customers by total spend, order frequency, and average order value to visualize the "drop-off" point for churned users.

вЂ‹Support Ticket Issues:
 Quantify tickets by category (e.g., delivery, product quality) and correlate them with churn status to identify "pain points."

вЂ‹Return/Refund Behaviour:
 Calculate the return-to-order ratio; a high ratio often acts as a leading indicator of churn.

вЂ‹Web/App Activity:
 Measure sessions per month and average time on site to identify disengagement patterns.

вЂ‹Campaign/Intervention History:
 Compare the churn rate of groups that received marketing/discounts vs. those that did not.

вЂ‹Churn Distribution:
Map the total churn rate across the database to understand the scale of the challenge.

вЂ‹2. Five Churn-Risk Hypotheses

вЂ‹Support-Friction Hypothesis: Customers who open more than two support tickets within a 30-day period have a significantly higher churn probability.

вЂ‹Engagement-Drop Hypothesis: A reduction of >50\% in monthly web/app sessions is a leading indicator of churn within the next 60 days.

вЂ‹Return-Dissatisfaction Hypothesis: Customers with a return rate exceeding 25\% are twice as likely to churn compared to the average user.

вЂ‹Incentive-Sensitivity Hypothesis: Customers who only purchase during promotional events (low monetary frequency outside of sales) exhibit higher churn rates when marketing interventions stop.

вЂ‹Tenure-Risk Hypothesis: First-time customers who do not place a second order within 45 days are at the highest risk of permanent attrition.
